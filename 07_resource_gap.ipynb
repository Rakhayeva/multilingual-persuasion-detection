## Finding 3: Resource Gap for Central Asian NLP

### Kazakh

Of the 11 datasets surveyed, 5 include Kazakh language data.
None of them provide persuasion technique annotation.
Available Kazakh resources cover sentiment analysis (KazSAnDRA),
named entity recognition (KazNERD), machine translation (KazParC),
and hate speech detection (Bolatbek et al., 2024).
The only Kazakh dataset with any manipulation-related annotation is
baiangali/fake_news, which covers fake/real news classification (Sambetbayeva et al., 2025),
a related but distinct task used as a proxy in this project.

**Conclusion:** A Kazakh persuasion-annotated corpus does not exist
in open access. This makes it impossible to train or properly evaluate
persuasion detection models for Kazakh at this stage.

### Russian

Four datasets include Russian with labels related to manipulation or
propaganda: SemEval-2023 Task 3 and SlavicNLP 2025 provide
persuasion technique annotation but require institutional affiliation
for access. Two openly available datasets exist, namely
tg-misinfo-data (Kyrychenko, 2022) and EUvsDisinfo (Leite et al., 2024),
but both use channel-level or binary disinformation labels rather than
span-level persuasion technique annotation.

**Conclusion:** Russian persuasion-annotated data exists but is not
openly accessible without institutional affiliation. Open Russian
datasets cover related tasks (propaganda detection, disinformation
classification) but not persuasion technique annotation at the text
span level, which is required for our research question.

### Implication for this project

This resource gap explains the experimental design of this project.
Training was conducted exclusively on English (PTC SemEval-2020),
the only openly accessible persuasion-annotated corpus in political
news domain. Kazakh and Russian were used as proxy test sets only,
with fake/real news labels as a proxy for manipulative content.
Cross-lingual transfer results (H1) reflect not only language
mismatch but also task mismatch caused by the absence of
properly annotated data for target languages.

### Implication for the proposed MSc thesis

This finding directly motivates the research agenda of the proposed
MSc thesis: *"Reliability and Interpretability of NLP Models for
Detecting Political Manipulation in Multilingual Contexts:
A Central Asian Perspective"*.

Addressing this gap requires one of the following directions:
(a) creating new annotated resources for Kazakh and Russian,
(b) developing annotation-efficient methods such as zero-shot or
few-shot transfer from English, or (c) systematic evaluation of
existing multilingual models on proxy tasks with explicit
documentation of task mismatch. All three directions are viable
research contributions for a thesis-level study.
